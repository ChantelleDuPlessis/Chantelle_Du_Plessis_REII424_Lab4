```python
#Task 0 - Plot
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np

style.use('ggplot')

x,y = np.loadtxt('punte.csv',
                 unpack=True,
                 delimiter = ',')

plt.plot(x,y, label='students',linewidth=5)

plt.title('Participation Mark for INGB315')
plt.ylabel('Percentage')
plt.xlabel('Students')
plt.legend()
plt.grid(True,color='b')
plt.savefig('Chantelle_DuPlessis_Task0_Plot.jpg')

plt.show()
```

```python
#Task 1 - Pie Chart
import matplotlib.pyplot as plt

figures = [73,3637,287,2142]
races = ['asian','white','coloured','african']
cols = ['m','r','b','y']

plt.pie(figures,
        labels=races,
        colors=cols,
        startangle=0,
        shadow= False,
        explode=(0.1,0,0,0),
        autopct='%1.0f%%')

plt.title('Racial composition of NWU staff as at May 2016')
plt.savefig('Chantelle_DuPlessis_Task1_PieChart_1.jpg')
plt.show()
```

```python
#Task 2 - Bar Chart
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np

style.use('ggplot')

x,y = np.loadtxt('punte.csv',
                 unpack=True,
                 delimiter = ',')

plt.bar(x,y, label='students',linewidth=3, align='center')

plt.title('Participation Mark for INGB315')
plt.ylabel('Percentage')
plt.xlabel('Students')
plt.legend()
plt.savefig('Chantelle_DuPlessis_Task2_BarChart.jpg')


plt.show()
```

```python
#Task 3 - Histogram
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np

style.use('ggplot')

x,y = np.loadtxt('punte.csv',
                 unpack=True,
                 delimiter = ',')

plt.hist(y,9, label='students',color = 'b', rwidth=0.8)

plt.title('Participation Mark for INGB315')
plt.ylabel('Students')
plt.xlabel('Percentages')
plt.legend()
plt.savefig('Chantelle_DuPlessis_Task3_Histogram.jpg')


plt.show()
```
